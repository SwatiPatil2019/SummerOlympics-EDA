Instead of checking yearwise records, like in which year where the Games were held - we will mark all cities on the World Map. 

````python
# import libraries
import folium
from IPython.display import HTML, display

# Make a data frame with dots to show on the map
data = pd.DataFrame({
'lat':[37,48,38,51,59,51,52,34,52,60,-37,41,35,23,48,45,55,37,41,33,-33,39],
'lon':[23,2,-90,-0.12,18,4,4,-118,13,24,144,12,139,-102,11,-73,37,126,2,-84,151,116],
'name':['Athens','Paris','St Louis', 'London', 'Stockholm', 'Antwerp',
       'Amsterdam', 'Los Angeles', 'Berlin', 'Helsinki',
       'Melbourne / Stockholm', 'Rome', 'Tokyo', 'Mexico', 'Munich',
       'Montreal', 'Moscow', 'Seoul', 'Barcelona', 'Atlanta', 'Sydney',
       'Beijing']
})

# Make an empty map
m = folium.Map(location=[20, 0], tiles="Mapbox Bright", zoom_start=2)
 
def years(tList):
    return [str(var) for var in tList]

# I can add marker one by one on the map
for i in range(0,len(data)):
    t = tempDict.loc[tempDict.City == data.iloc[i]['name']].Year
    popupStr = data.iloc[i]['name']+ ' [' + ','.join(years(t)) + ']'
    folium.Marker([data.iloc[i]['lat'], data.iloc[i]['lon']], popup=popupStr).add_to(m)
    
# Save it as html
m.save('SummerOlympicsCity.html')
````
